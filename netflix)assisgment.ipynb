```python
#task1
# Import necessary libraries
import pandas as pd

# Load the dataset
df = pd.read_csv("netflix_customer_churn.csv")  # replace with your local path if needed

# 1. Display the first 5 rows of the dataset
print(" First 5 rows of the dataset:")
display(df.head())

# 2. Check for missing/null values
print("\n Missing/Null Values in Each Column:")
print(df.isnull().sum())

# 3. Check for unique values in selected categorical columns
categorical_columns = ['gender', 'subscription_type', 'region', 'device', 'payment_method', 'favorite_genre']

print("\n Unique Values in Categorical Columns:")
for col in categorical_columns:
    print(f"{col}: {df[col].unique()}")
```

```python
# Task 2: Univariate Analysis

# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# -----------------------------
# 1. Distribution of Numerical Variables
# -----------------------------
plt.figure(figsize=(14, 10))

# Age Distribution
plt.subplot(2, 2, 1)
sns.histplot(df['age'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Age')

# Watch Hours Distribution
plt.subplot(2, 2, 2)
sns.histplot(df['watch_hours'], kde=True, bins=30, color='orange')
plt.title('Distribution of Watch Hours')

# Monthly Fee Distribution
plt.subplot(2, 2, 3)
sns.histplot(df['monthly_fee'], kde=True, bins=20, color='green')
plt.title('Distribution of Monthly Fee')

# Churned Distribution
plt.subplot(2, 2, 4)
sns.countplot(data=df, x='churned', palette='Set2')
plt.title('Churn Distribution (0 = No, 1 = Yes)')
plt.xlabel('Churned')

plt.tight_layout()
plt.show()
```

```python
# task3
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style="whitegrid")

# -----------------------------
# 1. Compare average watch_hours and monthly_fee 
#    across subscription_type, region, and device
# -----------------------------

group_cols = ['subscription_type', 'region', 'device']
metrics = ['watch_hours', 'monthly_fee']

plt.figure(figsize=(16, 18))

for i, col in enumerate(group_cols):
    for j, metric in enumerate(metrics):
        plt.subplot(3, 2, i*2 + j + 1)
        sns.barplot(data=df, x=col, y=metric, estimator='mean', ci='sd', palette='Set2')
        plt.title(f'Average {metric.replace("_", " ").title()} by {col.replace("_", " ").title()}')
        plt.xlabel(col.replace("_", " ").title())
        plt.ylabel(f'Average {metric.replace("_", " ").title()}')
        plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# -----------------------------
# 2. Plot average avg_watch_time_per_day by favorite_genre
# -----------------------------

plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='favorite_genre', y='avg_watch_time_per_day', estimator='mean', ci='sd', palette='coolwarm')
plt.title('Average Watch Time Per Day by Favorite Genre')
plt.xlabel('Favorite Genre')
plt.ylabel('Avg Watch Time Per Day (hrs)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# -----------------------------
# 3. Analyze churn rate based on gender, region, subscription_type, and payment_method
# -----------------------------

categorical_factors = ['gender', 'region', 'subscription_type', 'payment_method']

plt.figure(figsize=(16, 16))

for i, factor in enumerate(categorical_factors, 1):
    plt.subplot(2, 2, i)
    churn_rate = df.groupby(factor)['churned'].mean().reset_index()
    sns.barplot(data=churn_rate, x=factor, y='churned', palette='magma')
    plt.title(f'Churn Rate by {factor.replace("_", " ").title()}')
    plt.xlabel(factor.replace("_", " ").title())
    plt.ylabel('Churn Rate')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# task4
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="white")

# -----------------------------
# 1. Create a heatmap of correlation between numerical variables
# -----------------------------

# Selecting only numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).drop(columns=['churned'])

# Add churned column at the end for correlation check
correlation_data = numerical_cols.copy()
correlation_data['churned'] = df['churned']

# Compute correlation matrix
corr_matrix = correlation_data.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```
